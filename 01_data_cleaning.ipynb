# Data Cleaning
Clean up the text in the .txt files. Things to remove:
* Preface, table of content, publisher's notes
* HTML embeddings
* Blank lines
* Footnotes
* Chapter headings

```python
import os
import re
```

```python
DATA_PATH = './data/raw/'
os.listdir(DATA_PATH)
```

```python
OUTPUT_FOLDER = './data/cleaned/'
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

```python
CHAPTER_TOKEN = "[CHAPTER_BREAK]"
```

## 1. Sống mòn - Nam Cao

```python
with open(os.path.join(DATA_PATH, 'song_mon_nam_cao.txt'), 'r', encoding='utf-8') as file:
    text = file.readlines()

text = [line.strip() for line in text if line.strip() != ""]
```

```python
text[:6]
```

```python
# remove the intro (first 6 lines)
text = text[6:]
```

```python
text[:3]
```

```python
# remove the chapter headers and replace them with chapter break special token
# Phần 1, Phần 2, etc.
chapter_regex = r'^Phần\s+\d+'
text = [CHAPTER_TOKEN if re.match(chapter_regex, line.strip(), flags=re.IGNORECASE) 
    else line 
    for line in text]
```

```python
text[0:2]
```

```python
# remove end notes
text = text[:-2]
```

```python
# write to file
with open(os.path.join(OUTPUT_FOLDER, 'song_mon_nam_cao_stripped.txt'), 'w', encoding='utf-8') as file:
    file.writelines([line+'\n' for line in text])
```

## 2. Tắt đèn - Ngô Tất Tố

```python
with open(os.path.join(DATA_PATH, 'tat_den_ngo_tat_to.txt'), 'r', encoding='utf-8') as file:
    text = file.readlines()

text = [line.strip() for line in text if line.strip() != ""]
```

```python
# remove BOM markers
# \ufeff, ^h1, $h1
text = [line for line in text if line != '\ufeff']

text = [re.sub(r'[\^$]h\d+', '', line) for line in text]

# remove non-breaking space •\xa0
text = [line.replace('•', '').replace('\xa0', '') for line in text]
```

```python
# remove preface, introduction (first 39 lines)
text = text[39:]
```

```python
text[0:4]
```

```python
# remove the chapter headers and replace them with chapter break special token
chapter_regex = r'^CHƯƠNG\s+\d+'
text = [CHAPTER_TOKEN if re.match(chapter_regex, line.strip(), flags=re.IGNORECASE) 
    else line 
    for line in text]
```

```python
text[:4]
```

```python
# remove end note
text = text[:-2]
```

```python
# write to file
with open(os.path.join(OUTPUT_FOLDER, 'tat_den_ngo_tat_to_stripped.txt'), 'w', encoding='utf-8') as file:
    file.writelines([line+'\n' for line in text])
```

## 4. Đoạn tuyệt - Nhất Linh

```python
with open(os.path.join(DATA_PATH, 'doan_tuyet_nhat_linh.txt'), 'r', encoding='utf-8') as file:
    text = file.readlines()

text = [line.strip() for line in text if line.strip() != ""]
```

```python
# remove preface and table of content
text = text[37:]
```

```python
text[:5]
```

```python
# remove page header
# Nhất Linh
# Đoạn Tuyệt
text = [line for line in text if line.strip() not in ['Nhất Linh', 'Đoạn Tuyệt']]
```

```python
# remove section headers
# Phần thứ nhất, Phần thứ hai, etc.
text = [line for line in text if not re.match(r'^Phần\s+thứ\s+\w+', line.strip(), flags=re.IGNORECASE)]

# remove chapter headers
chapter_regex = r'^Chương\s+\d+'
text = [CHAPTER_TOKEN if re.match(chapter_regex, line.strip(), flags=re.IGNORECASE) 
    else line 
    for line in text]
# text = [line for line in text if not re.match(r'^Chương\s+\d+', line.strip(), flags=re.IGNORECASE)]
```

```python
text[:10]
```

```python
# remove end notes
text = text[:-6]
```

```python
# write to file
with open(os.path.join(OUTPUT_FOLDER, 'doan_tuyet_nhat_linh_stripped.txt'), 'w', encoding='utf-8') as file:
    file.writelines([line+'\n' for line in text])
```

## 5. Số Đỏ - Vũ Trọng Phụng

```python
with open(os.path.join(DATA_PATH, 'so_do_vu_trong_phung.txt'), 'r', encoding='utf-8') as file:
    text = file.readlines()

text = [line.strip() for line in text if line.strip() != ""]
```

```python
# remove table of content
text = text[22:]
```

```python
text[:5]
```

```python
# remove chapter headers
chapter_regex = r'^Chương\s+\d+'
text = [CHAPTER_TOKEN if re.match(chapter_regex, line.strip(), flags=re.IGNORECASE) 
    else line 
    for line in text]
```

```python
text[:10]
```

```python
# Remove in-line remarks with 1-2 digits, e.g., (1), (12)
text = [re.sub(r'\(\d{1,2}\)\s?', '', line).strip() for line in text]
```

```python
text[:10]
```

```python
# write to file
with open(os.path.join(OUTPUT_FOLDER, 'so_do_vu_trong_phung_stripped.txt'), 'w', encoding='utf-8') as file:
    file.writelines([line+'\n' for line in text])
```

## 6. Ngày mới  - Thạch Lam

```python
with open(os.path.join(DATA_PATH, 'ngay_moi_thach_lam.txt'), 'r', encoding='utf-8') as file:
    text = file.readlines()

text = [line.strip() for line in text if line.strip() != ""]
```

```python
text[:10]
```

```python
# remove the preface and the short stories 
# (this file is a compilation of Thach Lam's works but we only need the novel)

start_index = text.index('Ngày mới')
end_index = text.index('Ký sự Hà Nội băm sáu phố phường')

print(start_index, end_index)

text = text[start_index+1:end_index]
```

```python
text[:5]
```

```python
# remove section index
# Phần thứ nhất, Phần thứ hai, etc.
text = [line for line in text if not re.match(r'^Phần\s+thứ\s+\w+', line.strip(), flags=re.IGNORECASE)]

# remove chapter headers 
chapter_regex = r'^[IVXLCDM]+$'
text = [CHAPTER_TOKEN if re.match(chapter_regex, line.strip(), flags=re.IGNORECASE) 
    else line 
    for line in text]
```

```python
text[:5]
```

```python
# remove end notes
text = text[:-2]
```

```python
# write to file
with open(os.path.join(OUTPUT_FOLDER, 'ngay_moi_thach_lam_stripped.txt'), 'w', encoding='utf-8') as file:
    file.writelines([line+'\n' for line in text])
```
